## Feature Scaling:

**Scaling is necessary for distance-based algorithms such as KNN. This is to avoid higher weightage being assigned to data with a higher magnitude.**

**Using standard scaler we can scale down to unit variance.**

```python
DF= zoo.iloc[:,1:17]
```

```python
DF
```

### Using Standard Scaler:

```python
array = DF.values
array
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Normalization function:

stscaler = StandardScaler().fit(array)
x = stscaler.transform(array)
y = zoo['type']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.3,random_state=45)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## Building KNN Model:

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
```

```python
#Predicting on test data:

#predicting on test data set:
preds = model.predict(x_test)
```

```python
# getting the count of each category:

pd.Series(preds).value_counts() 
```

```python
#getting the 2 way table to understand the correct and wrong predictions:

pd.crosstab(y_test,preds)
```

```python
print("Accuracy", accuracy_score(y_test,preds)*100)
```

```python
model.score(x_train,y_train)
```

```python
print(classification_report(y_test,preds))
```

## Grid Search for Algorithm Tuning:

```python
n_neighbors = np.array(range(1,15))
param_grid = dict(n_neighbors=n_neighbors)
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator=model, param_grid=param_grid)
grid.fit(x, y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

## Visualizing the CV results:

```python
k_values = np.arange(1,25)
train_accuracy = []
test_accuracy = []

for i, k in enumerate(k_values):
    # k from 1 to 25(exclude)
    knn = KNeighborsClassifier(n_neighbors=k)
    # Fit with knn
    knn.fit(x_train,y_train)
    #train accuracy
    train_accuracy.append(knn.score(x_train, y_train))
    # test accuracy
    test_accuracy.append(knn.score(x_test, y_test))
# Plot
plt.figure(figsize=[13,8])
plt.plot(k_values, test_accuracy, label = 'Testing Accuracy')
plt.plot(k_values, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.title('-value VS Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.savefig('graph.png')
plt.show()
print("Best accuracy is {} with K = {}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
```
